### Plot Scatter plot using matplotlib
**Syntax** : plt.scatter(

                                      x,
                                      y,
                                      s=None,
                                      c=None,
                                      marker=None,
                                      cmap=None,
                                      norm=None,
                                      vmin=None,
                                      vmax=None,
                                      alpha=None,
                                      linewidths=None,
                                      verts=None, ### same work like – marker
                                      edgecolors=None,
                                      *,
                                      data=None,
                                      **kwargs,
                                      )

```python
plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.show()
```

```python
# Show scatter plot with title, xlabel and ylabel.
plt.scatter(x,y)
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
# Ploting Scatter plot with parameters
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
           edgecolors="g" )#verts="<"
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

### Plot Two Scatter Plot in one plot

```python
import matplotlib.pyplot as plt
import pandas as pd

# Load dataset
df = pd.read_csv("google_playstore_app.csv")

# Convert 'Installs' column to numeric (remove commas and '+')
df["Installs"] = df["Installs"].str.replace(",", "").str.replace("+", "").astype(int)

# Define x and y
x = df["Rating"]
y = df["Reviews"]

plt.figure(figsize=(16,9))

# Scatter plot 1: Rating vs Reviews
plt.scatter(x, y, c="r", marker="*", s=100, alpha=0.5, linewidths=2, edgecolors="g", label="Reviews")

# Scatter plot 2: Rating vs Installs
plt.scatter(x, df["Installs"], c="y", marker="o", s=100, alpha=0.5, linewidths=2, edgecolors="c", label="Installs")

# Titles and labels
plt.title("Google Play Store Apps Scatter Plot", fontsize=16)
plt.xlabel("Rating", fontsize=14)
plt.ylabel("Reviews & Installs", fontsize=14)

# Add legend to distinguish plots
plt.legend()

plt.show()
```
