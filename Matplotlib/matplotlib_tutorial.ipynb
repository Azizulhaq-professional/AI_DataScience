### What is python matplotlib?

Matplotlib is a Python library used to create graphs and plots.

You use it when you want to:

• Visualize data

• Understand trends

•Show results of your code

Most commonly used module: matplotlib.pyplot

Matplotlib is a 2D and 3D graph plotting python library. It also supports to create animations and images. This is an advanced part of python matplotlib.

Using python matplotlib, you can draw a different graph like below:

• Line plot

• Histogram

• Bar Chart

• Pie Chart

• Scatterplots

• Area Plot

• Error charts

• Power Spectra

• etc.

### Importing Matplotlib
For plotting graph, need to import matplotlib in the python file. Here is two way to import matplotlib given below.

```python
from matplotlib import pyplot as plt # importing pyplot module from matplotlib library
# or
import matplotlib.pyplot as plt #  importing pyplot module from
```

### 1. First Simple Plot (Line Graph)

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.show()
```

### Explanation

• x → values on X-axis

• y → values on Y-axis

• plt.plot() → draws the line

• plt.show() → displays the graph

### 2.Adding Labels & Title (Very Important)

```python
plt.plot(x, y)

plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Simple Line Graph")

plt.show()
```

Now your graph looks professional

### 3.Changing Line Style & Marker

```python
plt.plot(x, y, marker='o', linestyle='--')

plt.show()
```

### Common options:

• marker='o' → dots on points

• linestyle='--' → dashed line

• '-' → solid line

### 4.Multiple Lines in One Graph

```python
y2 = [5, 15, 20, 25]

plt.plot(x, y, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.legend()
plt.show()
```

### Explanation

• label → name of line

• plt.legend() → shows labels box

### 5.Bar Chart (Very Common)

```python
names = ['A', 'B', 'C']
values = [10, 30, 20]

plt.bar(names, values)
plt.title("Bar Chart Example")
plt.show()
```

### Used for:

• Comparison

• Categories

### 6.Scatter Plot (Dots Graph)

```python
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()
```

### Used for:

• Relationship between variables

• ML & Data Science

### 7.Histogram (Data Distribution)

```python
data = [1,2,2,3,3,3,4,4,4,4]

plt.hist(data)
plt.title("Histogram Example")
plt.show()
```

### Used for:

Frequency

Distribution analysis

### 8.Grid (Optional but Nice)

```python
plt.plot(x, y)
plt.grid(True)
plt.show()
```
