### Pandas merge()

### What is merge()?

merge() is used to combine two DataFrames based on common columns or indexes, similar to SQL JOINs.

Used in:

• Data analysis projects

• Finance & accounting datasets

• Logs correlation (CEH / SOC use cases)

### Basic Syntax

pd.merge(
    left,
    right,
    how='inner',
    on=None,
    left_on=None,
    right_on=None,
    left_index=False,
    right_index=False,
    sort=False,
    suffixes=('_x', '_y'),
    indicator=False,
    validate=None
)

### Step-by-Step Parameters Explanation
### 1.left

The first DataFrame.

pd.merge(df1, df2)

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Department': ['HR', 'IT', 'Finance']
})

# Merge the two DataFrames
result = pd.merge(df1, df2)

print(result)
```

### 2.right

The second DataFrame.

pd.merge(df1, df2)

These two DataFrames are mandatory.

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Department': ['HR', 'IT', 'Finance']
})

# Merge the two DataFrames
result = pd.merge(df1, df2)

print(result)
```

### 3.how – Type of Join (VERY IMPORTANT)
| Value   | Meaning                      |
|---------|------------------------------|
| `'inner'` | Common records only (default) |
| `'left'`  | All rows from left DF       |
| `'right'` | All rows from right DF      |
| `'outer'` | All rows from both          |

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Department': ['HR', 'IT', 'Finance']
})
```

```python
#  1.Inner Join
pd.merge(df1, df2, how='inner')
# Keeps only rows where keys (EmployeeID) exist in both DataFrames.
```

```python
# 2.Left Join
pd.merge(df1, df2, how='left')
#bKeeps all rows from df1, adds matches from df2. Missing values become NaN.
```

```python
# 3.Right Join
pd.merge(df1, df2, how='right')
# Keeps all rows from df2, adds matches from df1. Missing values become NaN.
```

```python
# 4.Outer Join
pd.merge(df1, df2, how='outer')
# Keeps all rows from both DataFrames, fills missing values with NaN.
```

### 4.on

Column name(s) common in both DataFrames.

pd.merge(df1, df2, on='EmployeeID')

Use only if column names are same.

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'EmployeeID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Department': ['HR', 'IT', 'Finance']
})

# Merge using EmployeeID as the key
result = pd.merge(df1, df2, on='EmployeeID')

print(result)
```

### 5.left_on & right_on

Used when column names are different.

e.g

pd.merge(
    df1,
    df2,
    left_on='Emp_ID',
    right_on='EmployeeID'
)

Use instead of on

```python
# Example
import pandas as pd

df1 = pd.DataFrame({
    'Emp_ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

df2 = pd.DataFrame({
    'EmployeeID': [1, 2, 4],
    'Department': ['HR', 'IT', 'Finance']
})

# Correct merge
result = pd.merge(df1, df2, left_on='Emp_ID', right_on='EmployeeID')
print(result)
```

### 6.left_index & right_index

Use index instead of columns.

e.g

pd.merge(
    df1,
    df2,
    left_index=True,
    right_index=True
)

Useful when index contains meaningful keys.

```python
# Example
import pandas as pd

# First DataFrame with custom index
df1 = pd.DataFrame({
    'Name': ['Ali', 'Sara', 'John']
}, index=[1, 2, 3])

# Second DataFrame with custom index
df2 = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance']
}, index=[1, 2, 4])

# Merge using index
result = pd.merge(df1, df2, left_index=True, right_index=True)

print(result)
```

### 7.sort

Sort result by join keys.

pd.merge(df1, df2, sort=True)

Default is False (faster).

```python
# Example

import pandas as pd

df1 = pd.DataFrame({
    'EmployeeID': [3, 1, 2],
    'Name': ['John', 'Ali', 'Sara']
})

df2 = pd.DataFrame({
    'EmployeeID': [2, 1, 3],
    'Department': ['IT', 'HR', 'Finance']
})

# Merge with sorting
result = pd.merge(df1, df2, on='EmployeeID', sort=True)

print(result)
```

### 8.suffixes

Resolve column name conflicts.

pd.merge(
    df1,
    df2,
    on='ID',
    suffixes=('_left', '_right')
)

Prevents overwriting column names.

```python
# Example

import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John'],
    'Salary': [40000, 50000, 60000]
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Aly', 'Sarah', 'Johnny'],   # same column name "Name"
    'Department': ['HR', 'IT', 'Finance']
})

# Merge with suffixes
result = pd.merge(df1, df2, on='ID', suffixes=('_left', '_right'))

print(result)
```

### 9.indicator

Adds a column showing source of each row.

pd.merge(df1, df2, indicator=True)

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Department': ['IT', 'Finance', 'HR']
})

# Merge with indicator
result = pd.merge(df1, df2, on='ID', how='outer', indicator=True)

print(result)
```

### 10.validate

Ensures merge integrity.

pd.merge(df1, df2, validate='one_to_one')

| Value        | Meaning                 |
|--------------|-------------------------|
| `one_to_one` | Unique keys both sides  |
| `one_to_many`| Left unique            |
| `many_to_one`| Right unique           |
| `many_to_many`| Default                |

```python
# Example
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Ali', 'Sara', 'John']
})

# Second DataFrame
df2 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Department': ['HR', 'IT', 'Finance']
})

# Merge with validation
result = pd.merge(df1, df2, on='ID', validate='one_to_one')

print(result)
```
