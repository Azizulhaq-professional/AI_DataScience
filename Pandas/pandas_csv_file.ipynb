If you want to read some specific rows of the dataset use nrows parameters

```python
# Read the CSV file 'Fortune_10.csv' into a DataFrame.
# The parameter nrows=1 tells pandas to only read the first row of data from the file.
# This is useful when you want to quickly preview the structure or sample of the dataset
# without loading the entire file into memory.
df = pd.read_csv('Fortune_10.csv', nrows = 1)
df
```

```python
df = pd.read_csv('Fortune_10.csv', nrows = 5)
df
```

```python
# Read the CSV file 'Fortune_10.csv' into a DataFrame.
# The parameter usecols=[0] tells pandas to only load the first column (index 0) from the file.
# This is useful when you want to work with just one column instead of the entire dataset.
df = pd.read_csv('Fortune_10.csv', usecols = [0])  # usecol use for specific columns parameters of the dataset
df
```

```python
# The parameter header=1 tells pandas to use the second row (row index 1) of the file as the column headers.
# By default, pandas uses the first row (row index 0) as headers.
# You use header=1 when the first row contains extra information, notes, or metadata,
# and the actual column names start from the second row.
df = pd.read_csv('Fortune_10.csv', header = 1) # Using Header Parameter
df
```

```python
# The parameter header=None tells pandas that the file does NOT have a header row (column names).
# In this case, pandas will automatically assign default numeric column labels (0, 1, 2, ...).
# This is useful when your CSV file only contains raw data without column names,
# or if you want to manually assign your own column names later.
df = pd.read_csv('Fortune_10.csv', header = None)
df
```

```python
import pandas as pd
df = pd.read_csv('Fortune_10.csv')
df
```

```python
df['Active'] = 'Yes' # Add a column with default values
df
```

```python
df['Active'] = df['Profit'] > 5000000 # Add a column with conditional logic.This will create a Boolean column (True or False).
df
```

```python
df['Active'] = df['Active'].replace({'Yes': True, 'No': False}) # Convert custom values to Boolean using true_values and false_values
df
```

```python
df.to_csv('Updated_Fortune_10.csv', index=False) # This will save the new file in your Colab environment.
```
