### Python Seaborn Tutorial
### 1.What is Seaborn?

Seaborn is a Python library used for statistical data visualization.

 • Built on top of Matplotlib
 • Works perfectly with Pandas DataFrames
• Less code, better-looking graphs

### 2.Install & Import Seaborn

```python
# Import
import seaborn as sns
import matplotlib.pyplot as plt
```

### Load Sample Dataset
Seaborn comes with built-in datasets for practice.

```python
data = sns.load_dataset("tips")
data.head()
```

### Dataset columns:

• total_bill

• tip

• sex

• smoker

• day

• time

• size

### 4.Line Plot (Trend Visualization)

```python
sns.lineplot(x="size", y="total_bill", data=data)
plt.show()
```

### Explanation

• x → X-axis column

• y → Y-axis column

• data → DataFrame

Used for trends

### 5.Scatter Plot (Most Important)

```python
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.show()
```

Shows relationship between two variables

### 6.Add Category (Hue)

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    data=data
)
plt.show()
```

### hue

• Splits data by category

• Automatically adds legend

Very powerful feature

### 7.Bar Plot (Average Values)

```python
sns.barplot(x="day", y="total_bill", data=data)
plt.show()
```

Seaborn automatically calculates mean

### 8.Count Plot (Frequency Count)

```python
sns.countplot(x="day", data=data)
plt.show()
```

Shows how many times each category appears

### 9.Box Plot (Outliers + Distribution)

```python
sns.boxplot(x="day", y="total_bill", data=data)
plt.show()
```

### Used in:

• Data analysis

• ML feature understanding

### 10.Histogram (Distribution)

```python
sns.histplot(data["total_bill"])
plt.show()
```

```python
# Or with KDE (curve):
sns.histplot(data["total_bill"], kde=True)
plt.show()
```

### 11.Heatmap (Correlation Matrix)

```python
corr = data.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.show()
```

### Very important for:

Machine Learning

Feature selection

### 12.Pair Plot (One Line = Full Analysis)

```python
sns.pairplot(data)
```

### Automatically:

• Scatter plots

• Histograms

•Correlations

Slightly slow on big datasets

### 13.Style & Theme (One Line Beauty)

```python
sns.set_style("darkgrid")
```

# Seaborn vs Matplotlib (Simple)

| Feature          | Matplotlib     | Seaborn       |
|------------------|----------------|---------------|
| Code length      | More           | Less          |
| Default look     | Basic          | Beautiful     |
| Pandas support   | Manual         | Automatic     |
| Statistical plots| Limited        | Excellent     |
