### What is a Seaborn Scatter Plot?

A scatter plot shows the relationship between two numerical variables using dots.

Each dot = one data point

### Used to:

• See patterns

• Check relationships

• Detect outliers

• Understand trends (ML & Data Analysis)

### 1.Import Required Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2.Load a Sample Dataset

Seaborn provides ready-to-use datasets.

```python
data = sns.load_dataset("tips")
data.head()
```

### Columns we’ll use:

• total_bill → numeric

• tip → numeric

• sex → category

• time → category

• size → numeric

### 3.Basic Scatter Plot (Most Simple)

```python
sns.scatterplot(x="total_bill", y="tip", data=data)
plt.show()
```

### Explanation (Line by Line)

• x="total_bill" → X-axis values

• y="tip" → Y-axis values

•data=data → DataFrame source

• Each dot = one customer

Shows how tip changes with total bill

### Add Category Using hue (Very Important)

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    data=data
)
plt.show()
```

### What hue Does

• Colors points by category

• Adds legend automatically

• Helps comparison

**Example:**

• Blue = Male

• Orange = Female

### 5.Change Marker Style Using style

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="time",
    data=data
)
plt.show()
```

### Meaning

• style="time" → Lunch vs Dinner

• Different marker shapes

### 6.Change Dot Size Using size

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    size="size",
    data=data
)
plt.show()
```

Bigger dots = larger group size

### 7.Control Size Range

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    size="size",
    sizes=(20, 200),
    data=data
)
plt.show()
```

• sizes=(min, max) controls dot scaling

### 8.Alpha (Transparency)

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    alpha=0.6,
    data=data
)
plt.show()
```

Useful when many points overlap

### 9.Add Title and Labels

```python
sns.scatterplot(x="total_bill", y="tip", data=data)

plt.title("Total Bill vs Tip")
plt.xlabel("Total Bill Amount")
plt.ylabel("Tip Amount")

plt.show()
```

### 10.Scatter Plot with Regression Line

```python
sns.regplot(x="total_bill", y="tip", data=data)
plt.show()
```

### Shows:

• Dots + trend line

• Used in ML & statistics

### Important Parameters Summary

| Parameter | Purpose |
|-----------|---------|
| x | X-axis variable |
| y | Y-axis variable |
| data | DataFrame |
| hue | Color grouping |
| style | Marker grouping |
| size | Dot size |
| sizes | Size range |
| alpha | Transparency |
