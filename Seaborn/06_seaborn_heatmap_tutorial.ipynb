### What is a Heatmap?

A heatmap is a data visualization technique that shows values in a matrix (2D data) using colors.

• Each cell in the matrix is colored according to its value.

• Darker or lighter shades represent higher or lower values.

• It’s especially useful for visualizing correlations, confusion matrices, or tabular data.

### Seaborn and sns.heatmap()

Seaborn is a Python library built on top of Matplotlib that makes statistical visualizations easier.
sns.heatmap() is the function used to create heatmaps.

### Step-by-Step Explanation

### 1. Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

### 2. Prepare Data

Heatmaps require 2D data (matrix-like).

Example: Create a random matrix using NumPy.

```python
data = np.random.rand(5, 5)   # 5x5 matrix with random values
```

### 3. Basic Heatmap


```python
sns.heatmap(data)
plt.show()
```

This will display a simple heatmap where colors represent values.

### 4. Add Annotations

You can show the actual numbers inside each cell:

```python
sns.heatmap(data, annot=True)
plt.show()
```

Each cell now displays its numeric value.

### 5. Control Color Scheme

You can change the color palette using cmap:

```python
sns.heatmap(data, annot=True, cmap="coolwarm")
plt.show()
```

"coolwarm", "viridis", "YlGnBu" are common colormaps.

### 6. Adjust Value Range

You can fix the color scale using vmin and vmax:

```python
sns.heatmap(data, annot=True, cmap="YlGnBu", vmin=0, vmax=1)
plt.show()
```

Ensures consistent coloring across multiple heatmaps.

### 7. Hide Color Bar

If you don’t want the side color bar:

```python
sns.heatmap(data, annot=True, cmap="YlGnBu", cbar=False)
plt.show()
```

```python
import pandas as pd
df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
```

### Step 2: Prepare the Data

We want to analyze correlations between years (how emissions in one year relate to another).

```python
# Drop non-numeric columns (Country Name, Code, Indicator Name, Indicator Code)
numeric_df = df.drop(columns=["Country Name","Country Code","Indicator Name","Indicator Code"])

# Compute correlation matrix
corr = numeric_df.corr()
```

corr is now a 15×15 matrix showing correlations between years 2000–2014.

### Step 3: Basic Heatmap

```python
sns.heatmap(corr)
plt.show()
```

This shows a simple heatmap of correlations.

## Step 4: Heatmap with Parameters Explained

```python
sns.heatmap(
    corr,                # The correlation matrix (2D data)
    annot=True,          # Show numeric correlation values inside cells
    fmt=".2f",           # Format numbers to 2 decimal places
    cmap="coolwarm",     # Color scheme (blue = negative, red = positive)
    linewidths=0.5,      # Add lines between cells
    linecolor="white",   # Color of the lines
    cbar=True,           # Show color bar (legend for values)
    square=True,         # Make cells square-shaped
    xticklabels=True,    # Show column labels (years)
    yticklabels=True     # Show row labels (years)
)
plt.title("Correlation Matrix of CO₂ Emissions (2000–2014)")
plt.show()
```

### Parameter-by-Parameter Explanation

| Parameter | Meaning | Example |
|:----------|:--------|:--------|
| data | 2D dataset (matrix or DataFrame) | corr |
| annot | If True, shows values inside cells | annot=True |
| fmt | Format of numbers | fmt=".2f" → 2 decimals |
| cmap | Color palette | "coolwarm", "viridis", "YlGnBu" |
| linewidths | Thickness of lines between cells | 0.5 |
| linecolor | Color of those lines | "white" |
| cbar | Show/hide color bar | True or False |
| square | Force square cells | True |
| xticklabels / yticklabels | Show/hide labels | True or False |
| vmin / vmax | Fix min/max values for color scale | vmin=-1, vmax=1 |

### Step 5: Interpretation

• Correlation close to 1 (red) → emissions in two years rise/fall together.

• Correlation close to 0 (white) → weak relationship.

• Correlation close to -1 (blue) → inverse relationship (rare in this dataset).

In my dataset, you’ll likely see strong positive correlations between consecutive years (e.g., 2005 vs 2006), because CO₂ emissions usually change gradually.

### Example Insight

If the heatmap shows:

• 2000 vs 2001 → 0.99 (almost identical trend).

• 2000 vs 2014 → 0.85 (still strongly correlated, but less).

This means emissions patterns are consistent across years, but some countries diverged over time.
