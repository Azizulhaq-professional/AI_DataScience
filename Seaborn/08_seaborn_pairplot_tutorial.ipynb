### Seaborn pairplot() Tutorial

### 1.What is a Pairplot? (In Simple Words)

A pairplot:

• Shows relationships between ALL numerical columns

• Uses:

   • Scatter plots for relationships

   • Histograms / KDE on the diagonal

 With one line of code, you get many plots.

### 2.When Do We Use Pairplot?

• Exploratory Data Analysis (EDA)

• Finding relationships

• Detecting patterns & outliers

• Before building ML models

### 3.Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

### Load a Dataset (Example)

We’ll use the Titanic dataset (you already know this one).

```python
import seaborn as sns
df = sns.load_dataset("titanic")
df.head()
```

### 5.Basic Pairplot (Simplest Form)

```python
sns.pairplot(df)
```

### What Happens Here?

• Every numeric column is plotted against every other numeric column

• Diagonal → histogram of each column

• Off-diagonal → scatter plots

This can be slow for large datasets.

### 6.Pairplot with hue (MOST IMPORTANT PARAMETER)

```python
sns.pairplot(df, hue="survived")
```

### hue

• Separates data by category

• Colors points by class

• Adds legend automatically

This helps you see how survival differs.

### 7.Control Diagonal Plot Type (diag_kind)

```python
# Default → Histogram
sns.pairplot(df, diag_kind="hist")
```

```python
# sns.pairplot(df, diag_kind="hist")
sns.pairplot(df, diag_kind="kde")
```

KDE gives a smooth distribution curve.

### 8.Control Off-Diagonal Plot Type (kind)

```python
# Default → Scatter
sns.pairplot(df, kind="scatter")
```

```python
# Regression Line
sns.pairplot(df, kind="reg")
```

reg adds trend lines (very useful in ML).

### 9.Select Specific Columns (vars)

```python
sns.pairplot(
    df,
    vars=["age", "fare", "sibsp"],
    hue="survived"
)
```

### Best practice:

• Use only relevant columns

• Makes plots faster & cleaner

### 10.Change Marker Style (markers)

```python
sns.pairplot(
    df,
    hue="survived",
    markers=["o", "s"]
)
```

• o → circle

• s → square

### 11.Adjust Plot Size (height, aspect)

```python
sns.pairplot(
    df,
    hue="survived",
    height=2.5,
    aspect=1.2
)
```

Helps when plots look too small or crowded.

### 12.Transparency (plot_kws)

```python
sns.pairplot(
    df,
    hue="survived",
    plot_kws={"alpha": 0.6}
)
```

Useful when points overlap.

### 13.Pairplot with Given CSV

```python
numeric_df = df.select_dtypes(include="number")

sns.pairplot(numeric_df)
```

### Best practice:

• Pairplot only numeric data

• Avoid errors & clutter

### Important Parameters Summary

| Parameter | Meaning |
|:----------|:--------|
| data | Dataset |
| hue | Category separation |
| vars | Select columns |
| kind | Scatter or regression |
| diag_kind | Hist or KDE |
| markers | Point shapes |
| height | Size of plots |
| aspect | Width ratio |
| plot_kws | Extra styling |
